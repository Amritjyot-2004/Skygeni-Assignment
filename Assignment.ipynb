Q2: Which industry in the organization has the highest renewal rate?<br>
Process: Following steps are followed to obtain the answer:<br>
1) Dataframes icd and subinfo are joined using inner join and the joined dataframe is assigned to the variable inner_join.<br>
2) Total number of renewals for each industry is calculated by grouping the inner_join dataframe on industry column and using count function on renewed column. The obtained answer is assigned to variable total_renewals.<br>
3) Number of true renewals for each industry is calculated by grouping the inner_join dataframe on industry column and filtering out the rows hanving value of renewed column as false and then using count function on industry column itself. The obtained answer is assigned to variable renewals_accepted.<br>
4) The renewal rates for each industry are calculated by dividing each value of renewals_accepted from that of total renewals and rounding them to 2 decimal places.<br>
5) A new dataframe df is created with columns as industry, total renewals, renewals accepted, renewals rejected and renewal rate which is further used to visualize the renewal rates for each industry on bar plot and number of renewals accepted and rejected on double bar plot.<br>
6) Finally the maximum renewal rate is calculated by using np.max() function on renewal rate column of df and the industry with highest renewal rate is obtained.

```python
#Inner join for icd and subinfo dataframes on client id
inner_join = pd.merge(icd,subinfo, on="client_id", how = 'inner')
inner_join.head()
```

```python
#Grouping inner_join dataframe on industry column and finding total number of renewals for each industry by using count() function
total_renewals = inner_join.groupby('industry')['renewed'].count()
total_renewals
```

```python
#Grouping inner_join dataframe on industry column and filtering out rows with value of renewed = false. 
# Also finding total number of renewals for each industry by using count() function.
df = inner_join.groupby('industry').apply(lambda group: group[group['renewed']== True]).reset_index(drop=True)
renewals_accepted = df['industry'].value_counts()
renewals_accepted
```

Since the indices for industry names in total_renewals.index list and renewals_accepted.index list were not in same position,so a new list ra was created which contains the count of accepted renewals for each industry in the same position corresponsing to that of total_renewals.index

```python
ra = [renewals_accepted[i] for i in total_renewals.index]   #List containing the values of renewals_accepted
rr = [total_renewals[i]-renewals_accepted[i] for i in total_renewals.index] #List containing values of renewals rejected
rrate = [round(renewals_accepted[i]/total_renewals[i],2) for i in total_renewals.index] #List containing the renewal rates
df1 = pd.DataFrame({"Industry":total_renewals.index, "Total Renewals":total_renewals.values, "Renewals Accepted": ra, "Renewals Rejected":rr, "Renewal Rate":rrate})  #Creating the dataframe df
df1
```

```python
#Visualizing renewals accepted and rejected using double bar plot

#Bar Positions
x = np.arange(len(df1['Industry']))     #Positions for x labels
width = 0.4                             #Width of the bars

#Plotting
plt.figure(figsize=(8, 5))
plt.bar(x - width/2, df1['Renewals Accepted'], width, label='True', color='green')
plt.bar(x + width/2, df1['Renewals Rejected'], width, label='False', color='red')
plt.xlabel('Industry')
plt.ylabel('Number of Renewals')
plt.title('Number of Renewals Accepted/Rejected for each Industry')
plt.xticks(x, df1['Industry'])  # Set x-ticks to category names
plt.legend()        #For displaying true and false labels on top right corner of figure
plt.show()
```

```python
#Visualizing the renewal rates using bar plot
plt.figure(figsize=(10,6))
plt.bar(df1['Industry'],df1['Renewal Rate'], color=['red','green','cyan','magenta','black'])
plt.xlabel('Industry')
plt.ylabel('Renewal Rate')
plt.title("Industry Wise Renewal Rate")
plt.show()
```

```python
#Calculating the max renewal rate and industry corresponding to it.
maxRRate = np.max(df1['Renewal Rate'])
res = ', '.join(df1[df1['Renewal Rate']==maxRRate]['Industry'].values)
#Final answer to Q2
print(f"Q: Which industry has highest renewal rate?")
print(f"{res} has highest renewal rate of {maxRRate}")
```

Q3: What was the average inflation rate when their subscriptions were renewed?<br>
Process: The following steps are followed to obtain the answer<br>
1) Every value of the payment_date column of payinfo dataframe is converted into a datetime object. Same step was repeated for start_date and end_date columns of fi dataframe. This was done so as to check if payment_date lie in the range of start_date and end_date.<br>
2) The distribution of inflation rates for every start_date of fi dataframe is visualized using combo chart of bar and line plot.<br>
3) Dataframes payinfo and df are joined using inner join on client_id column so as to take into account the payment_dates of those clients for which the subscriptions are renewed.<br>
4) The inflation rates associated with the payment_dates that lie in the range of any start_date, end_date pair of fi dataframe was obtained in the list inf_rate and then a new column named inflation rate is added in the dataframe df containing same values as that of list inf_rate.<br>
5) Finally the average of the inflation rate column rounded off to 2 decimal places is calculated using np.mean() function.

```python
#Converting every value of payment_date of payinfo dataframe to datetime object
from datetime import datetime
for i in range(len(payinfo['payment_date'])):
    if(payinfo['payment_date'][i][1]=='/'or payinfo['payment_date'][i][2]=='/'):
        payinfo['payment_date'][i] = datetime.strptime(payinfo['payment_date'][i], "%m/%d/%Y")  #Parsing the string into datetime object using same format as that of string
    else:
        payinfo['payment_date'][i] = datetime.strptime(payinfo['payment_date'][i], "%d-%m-%Y")  #Parsing the string into datetime object using same format as that of string
payinfo.head()
```

```python
#Visualizing the inflation rate for each start date using combo chart
plt.figure(figsize=(10,80))
fig, ax = plt.subplots()        # Create the figure and axis

# Plot bar chart
ax.bar(fi['start_date'], fi['inflation_rate'], color = 'blue')
ax.set_ylabel('Inflation Rate')
ax.set_xlabel('Start Date')

#Setting labels for x axis
ax.set_xticks(range(len(fi['start_date'])))
ax.set_xticklabels(fi['start_date'], rotation=90)

#Plot the line chart
ax1 = ax.twinx()    #Creating second y axis
ax1.plot(fi['start_date'], fi['inflation_rate'], color='red', marker='o')
ax1.set_ylabel('Inflation Rate')

plt.title("Combo Chart for Inflation Rate")
plt.show()
```

```python
#Converting every value of payment_date of payinfo dataframe to datetime object
for i in range(21):
    fi['start_date'][i] = datetime.strptime(fi['start_date'][i], "%d-%m-%Y")    #Parsing the string into datetime object using same format as that of string
    fi['end_date'][i] = datetime.strptime(fi['end_date'][i],"%d-%m-%Y")         #Parsing the string into datetime object using same format as that of string
fi.head()
```

```python
#Inner join of payinfo and df
df2 = pd.merge(payinfo,df,on='client_id',how = 'inner')
df2
```

```python
#Obtaining the inflation rates associated with the payment_dates lying in range of any start_date - end_date pairs.
inf_rate = []
for i in df2['payment_date']:
    for j in range(21):
        if fi['start_date'][j]<=i<=fi['end_date'][j]:
            inf_rate.append(fi['inflation_rate'][j])
df2['inflation_rate'] = inf_rate
df2.head()
```

```python
#Final answer of Q3
print(f"Q: What was the average inflation rate when their subscriptions were renewed?")
print(f"The average inflation rate when their subscriptions were renewed was {round(np.mean(df2['inflation_rate']),2)}")
```

Q4:What is the median amount paid each year for all payment methods?<br>
Process: Following steps are followed to obtain the answer:<br>
1) From all values of payement_date column of payinfo dataframe, years are extracted and stored in Years list. this was done to calculate median amount for each year.<br>
2) A new column named Years is added to the dataframe payinfo using years list.<br>
3) The median amount for each year for all payment methods is calculated by grouping the dataframe payinfo by columns Year and payment_method and applying median function on amount_paid column.<br>
4) The grouped dataframe is converted to dataframe with columns as Years, Payment Methods and Median Amounts.<br>
5) The obtained dataframe is used to visualize the distribution of the total median amount per year over each payment method using pie charts.<br>
6) Finally the dataframe is displayed as answer.<br>

```python
#Extracting years from payment_dates
Years = [payinfo['payment_date'][i].year for i in range(len(payinfo))]
Years[0:10]
```

```python
#Adding column year in payinfo dataframe
payinfo['Year'] = Years
payinfo.head()
```

```python
#Grouping payinfo dataframe by columns Year and payment_method and applying median function on amount_paid column
df4 = payinfo.groupby(['Year','payment_method'])['amount_paid'].median()
```

```python
#Creating the dataframe for visualizing
years = [i[0] for i in df4.index]
pay_meth = [i[1] for i in df4.index]
med_amt = df4.values
df4 = pd.DataFrame({"Years":years,"Payment Methods":pay_meth,"Median Amount":med_amt})
df4.head()
```

```python
#Visualizing the dataframe using piecharts
year = df4['Years'].unique()        #Gives unique values for years column

for yr in year:
    df_year = df4[df4['Years'] == yr]  #Returns a dataframe of rows where year column as value = yr
    total = round(df_year['Median Amount'].sum(),2)     #Calculates total median amount for year yr
    plt.figure(figsize=(6, 6))
    plt.pie(
        df_year['Median Amount'], 
        labels=df_year['Payment Methods'], 
        autopct='%1.1f%%',                  #Format to show amounts as percentage on pie chart               
        startangle=90, 
        colors=['Red', 'Green', 'Blue']
    )
    plt.title(f'Payment Method Distribution in {yr}\nTotal Median Amount:{total}')
    plt.show()
```
